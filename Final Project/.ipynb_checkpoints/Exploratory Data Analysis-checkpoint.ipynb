## 2. Basic stats

```python
df_shooting=pd.read_csv('NYPD_Shooting_Incident_Data__Historic_.csv')
print(df_shooting.shape)
print(df_shooting.columns)
```

```python
print(df_shooting.info())
```

```python
columns_shooting=['INCIDENT_KEY', 'OCCUR_DATE', 'OCCUR_TIME', 'BORO', 'PRECINCT',
       'JURISDICTION_CODE', 'LOCATION_DESC', 'STATISTICAL_MURDER_FLAG',
       'PERP_AGE_GROUP', 'PERP_SEX', 'PERP_RACE', 'VIC_AGE_GROUP', 'VIC_SEX',
       'VIC_RACE','Latitude', 'Longitude',
       'Lon_Lat'] # Keep these columns
df_shooting=df_shooting[columns_shooting]
print(df_shooting.shape)
```

```python
print(df_shooting['BORO'].value_counts(),"\n"*2,
      df_shooting['LOCATION_DESC'].value_counts(),"\n"*2,
     df_shooting['STATISTICAL_MURDER_FLAG'].value_counts(),"\n"*2,
     df_shooting['PERP_AGE_GROUP'].value_counts())
```

```python
# Define a list of values to exclude
exclude_values = ['UNKNOWN', '1020', '224', '940']

# Filter out values from the LOCATION_DESC column that are in the exclude list
df_filtered = df_shooting.loc[~df_shooting['LOCATION_DESC'].isin(exclude_values)]

# Define the order of categories based on shooting incident count
top_categories = df_filtered['LOCATION_DESC'].value_counts().nlargest(10).index

# Convert LOCATION_DESC column to categorical data type with specified order
df_filtered['LOCATION_DESC'] = pd.Categorical(df_filtered['LOCATION_DESC'], categories=top_categories)

# Calculate shooting incident count for each location description category
incident_count = df_filtered['LOCATION_DESC'].value_counts()

# Create horizontal bar plot using Seaborn with the ordered groups
sns.barplot(x=incident_count.values, y=incident_count.index, color='mediumturquoise')

# Set plot title and axis labels
plt.title('Shooting Incident Count by Location Description', fontsize=14)
plt.xlabel('Incident Count', fontsize=12)
plt.ylabel('Location Description', fontsize=12)

# Remove top and right spines
sns.despine()

# Add value labels to bars
for index, value in enumerate(incident_count):
    plt.text(value, index, str(value), ha='left', va='center', fontweight='bold', color='black')

# Show plot
plt.show()
```

```python
# Filter out 'UNKNOWN' values from the VIC_AGE_GROUP column
df_filtered = df_shooting.loc[df_shooting['VIC_AGE_GROUP'] != 'UNKNOWN']

# Calculate murder ratios for each victim age group
murder_ratios = df_filtered.groupby('VIC_AGE_GROUP')['STATISTICAL_MURDER_FLAG'].mean()

# Sort the murder ratios by ascending order and get the corresponding VIC_AGE_GROUP values
ordered_groups = murder_ratios.sort_values().index.tolist()

# Create bar plot using Seaborn with the ordered groups
sns.barplot(x='VIC_AGE_GROUP', y='STATISTICAL_MURDER_FLAG', data=df_filtered, order=ordered_groups, color='mediumturquoise')

# Set plot title and axis labels
plt.title('Fatality Ratio by Victim Age Group')
plt.xlabel('Victim Age Group')
plt.ylabel('Murder Ratio')

# Show plot
plt.show()
```

```python
# Define a list of values to exclude
exclude_values = ['UNKNOWN', '1020', '224', '940']

# Filter out values from the PERP_AGE_GROUP column that are not in the exclude list
df_filtered = df_shooting.loc[~df_shooting['PERP_AGE_GROUP'].isin(exclude_values)]

# Define the order of categories based on age group
age_order = ['<18', '18-24', '25-44', '45-64', '65+']

# Convert PERP_AGE_GROUP column to categorical data type with specified order
df_filtered['PERP_AGE_GROUP'] = pd.Categorical(df_filtered['PERP_AGE_GROUP'], categories=age_order)

# Calculate shooting incident count for each perpetrator age group
incident_count = df_filtered.groupby('PERP_AGE_GROUP')['INCIDENT_KEY'].count()

# Create bar plot using Seaborn with the ordered groups
sns.barplot(x=incident_count.index, y=incident_count.values, color='mediumturquoise')

# Set plot title and axis labels
plt.title('Shooting Incident Count by Perpetrator Age Group', fontsize=14)
plt.xlabel('Perpetrator Age Group', fontsize=12)
plt.ylabel('Incident Count', fontsize=12)

# Remove top and right spines
sns.despine()

# Add value labels to bars
for index, value in enumerate(incident_count):
    plt.text(index, value, str(value), ha='center', va='bottom', fontweight='bold', color='black')

# Show plot
plt.show()
```

## 3. Data Analysis

```python
# Import the required libraries
import datetime
import folium
from folium.plugins import HeatMapWithTime
geo={"lat": 40.730610 , "lon": -73.935242 }


# Filter the dataframe to get the required data for heat map
#heat_df = df[(df['Category'] == 'DRUG/NARCOTIC') & (df['Date'] < cut_off) & (df['Date'] > cut_off2)]
heat_df=df_shooting

# Add a column for count
heat_df['count'] = 1

# Create a list of day-wise data
df_day_list = []

# Format the dates and create a list of unique dates
#heat_df["DateFormatted"] = heat_df["Date"].dt.strftime("%B %-d%a")
heat_df["Date"] = pd.to_datetime(heat_df['OCCUR_DATE'], format='%m/%d/%Y')
#heat_df["DateFormatted"] = heat_df["Date"].dt.strftime("%B %-d%a")
heat_df['dayy']=heat_df['Date'].dt.year

heat_df['Date'] = heat_df['Date'].astype(str)
dates = list(set(heat_df["dayy"]))
indexx = sorted(dates)
#date_objs = [datetime.datetime.strptime(date, "%Y-%m-%d %H:%M:%S") for date in dates_sorted]
#indexx = [date_obj.strftime("%A %d. %B %Y") for date_obj in date_objs]

# Group the data by day, location and count
for day in heat_df.dayy.sort_values().unique():
    df_day_list.append(heat_df.loc[heat_df.dayy == day, ['Latitude', 'Longitude', 'count']].groupby(['Latitude', 'Longitude']).sum().reset_index().values.tolist())

# Create a folium map object
map_hooray = folium.Map(location=[geo['lat'], geo['lon']], zoom_start=11, tiles="Stamen Toner")


# Convert the latitude and longitude values to floats
heat_df['Y'] = heat_df['Latitude'].astype(float)
heat_df['X'] = heat_df['Longitude'].astype(float)

# Filter the data for NaN values and create a list of location data
heat_df = heat_df[['Y', 'X', 'dayy']]
heat_df = heat_df.dropna(axis=0, subset=['Y', 'X', 'dayy'])
heat_data = [[row['Y'], row['X']] for index, row in heat_df.iterrows()]

# Add the heat map with time to the folium map object
HeatMapWithTime(df_day_list, name='test', auto_play=False, max_opacity=1,index=indexx).add_to(map_hooray)

map_hooray

# Save the map to a file
#map_hooray.save('my_map.html')
```
